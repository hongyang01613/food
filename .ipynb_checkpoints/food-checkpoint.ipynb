```python
def select_food(data):
    data_index = data.index      # food list 
    food = choice(data_index)    # random chooce a food
    return food
```

```python
def create_cookbook(data):
    cookbook = {}                    
    quantity_cookbook_food = 40      #food variety 
    gram_food = 100                  # every food weight, unit(g)
    for i in range(quantity_cookbook_food):        
        cookbook[select_food(data)] = gram_food
    return cookbook
```

```python
ingredient = ['能量', '水分', '蛋白质', '脂肪', '膳食纤维', '碳水化物', '视黄醇当量', '硫胺素(VB1)', '核黄素(VB2)',
             '尼克酸(烟酸VPP)', '维生素E', '钠', '钙', '铁', '抗坏血酸(VC)', '胆固醇']
dict_ingredient = {'能量': 0, '水分': 0, '蛋白质': 0, '脂肪': 0, '膳食纤维': 0, '碳水化物': 0,
                   '视黄醇当量': 0, '硫胺素(VB1)': 0, '核黄素(VB2)': 0, '尼克酸(烟酸VPP)': 0, 
                   '维生素E': 0, '钠': 0, '钙': 0, '铁': 0, '抗坏血酸(VC)': 0, '胆固醇': 0}

#food*可食部分*数量/100 * ingredient
for ingre_ in list(dict_ingredient):
    print(ingre_, ' ', end = '')
print('')
cookbook = create_cookbook(data)
for name in list(cookbook):

    energy = 0
    protein = 0
    fat = 0
    carbohydrate = 0
    dietary_fiber = 0
    retinoflavone_equivalent = 0
    v_b1 = 0
    v_b2 = 0
    v_c = 0
    Ca = 0
    Fe = 0

    n_char = len(name)
    n = 25 - 2 * n_char
    print(name, n * ' ', end = '')
    
    for ingre_ in list(dict_ingredient):
        quantity_ingredient = data.loc[name]['可食部分'] / 100 * cookbook[name] / 100 * data.loc[name][ingre_]
        if np.isnan(quantity_ingredient):
            pass
        else:
            dict_ingredient[ingre_] += quantity_ingredient
        print("{:>6.2f} ".format(quantity_ingredient), end = '')
    print('')
print('')    
print('total:                    ', end='')
for ingre_ in list(dict_ingredient):
    print("{:>6.1f} ".format(dict_ingredient[ingre_]), end = '')
```
